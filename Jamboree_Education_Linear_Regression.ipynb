```python
def plot_scatter_relationships(df, target_column, predictors):
    """
    Plots scatter plots with regression lines for continuous variables against the target variable.

    Parameters:
        df (pd.DataFrame): The dataset.
        target_column (str): The column name of the target variable.
        predictors (list): List of predictor column names to plot against the target variable.
    """
    num_predictors = len(predictors)
    rows = (num_predictors + 1) // 3

    fig, axes = plt.subplots(rows, 3, figsize=(14, 5 * rows))
    axes = axes.flatten()

    for i, predictor in enumerate(predictors):
        sns.regplot(ax=axes[i], x=df[predictor], y=df[target_column], line_kws={"color": "red"}, scatter_kws={"alpha": 0.5})
        axes[i].set_title(f"{predictor} vs {target_column}")
        axes[i].set_xlabel(predictor)
        axes[i].set_ylabel(target_column)

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        axes[j].set_visible(False)

    plt.tight_layout()
    plt.show()


predictors = ["GRE Score", "TOEFL Score", "CGPA"]
plot_scatter_relationships(df=df, target_column="Chance of Admit", predictors=predictors)
```

```python
# Function to remove outliers using IQR
def remove_outliers_iqr(df, numerical_columns):
    plt.figure(figsize=(15, 10))

    for i, column in enumerate(numerical_columns, 1):
        # Calculate Q1 (25th percentile) and Q3 (75th percentile)
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1

        # Calculate lower and upper bounds
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Clip the column data to the calculated bounds
        df[column] = np.clip(df[column], lower_bound, upper_bound)

        # Plot boxplot after outlier removal
        plt.subplot(3, 3, i)
        sns.boxplot(df[column])
        plt.title(f'Boxplot of {column} (After Outlier Removal)')

    plt.tight_layout()
    plt.show()

    # Return the modified DataFrame
    return df


# Apply outlier removal using IQR
df = remove_outliers_iqr(df, ["LOR", "Chance of Admit"])
```

## **Multicollinearity**

```python
# Calculate VIF for each feature
def calculate_vif(X):
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    return vif_data
```

```python
def drop_high_vif_features(X, vif_threshold=5):
    """
    Drop variables with VIF greater than the specified threshold iteratively.

    Parameters:
    X (pd.DataFrame): DataFrame of predictors.
    vif_threshold (float): The threshold VIF value beyond which features will be dropped.

    Returns:
    pd.DataFrame: DataFrame with features having VIF <= threshold.
    """
    # Ensure the constant term is included for VIF calculation
    X_vif = X.copy()

    # Calculate the initial VIF
    vif_data = calculate_vif(X_vif)
    print(f"Initial VIF values:\n{vif_data}\n")

    # Drop variables with VIF > threshold iteratively
    while vif_data["VIF"].max() > vif_threshold:
        # Find the feature with the highest VIF
        feature_to_drop = vif_data.sort_values("VIF", ascending=False).iloc[0]["Feature"]

        # Drop the feature from the dataset
        X_vif = X_vif.drop(columns=[feature_to_drop])

        # Recalculate the VIFs after dropping the feature
        vif_data = calculate_vif(X_vif)

        print(f"Dropped feature: {feature_to_drop}")
        print(f"Updated VIF values:\n{vif_data}\n")

    # Return the DataFrame with reduced features
    return X_vif
```

### **Perform the Goldfeld-Quandt test**

> **Formulate Hypotheses**

  - **Null Hypothesis (H₀):** Residuals are homoscedastic (constant variance).
  - **Alternate Hypothesis (H₁):** Residuals are heteroscedastic (variance changes with predicted values).

> **Set a Significance Level**

  - Choose a significance level (**α**) = 0.05.
  - If the **p-value > 0.05**, we fail to reject the null hypothesis, indicating homoscedasticity.


```python
# Perform the Goldfeld-Quandt test
test_statistic, p_value, _ = het_goldfeldquandt(y_train, X_train_scaled_reduced)

# Print the results
print(f"Goldfeld-Quandt Test Statistic: {test_statistic}")
print(f"p-value: {p_value}")
```

```python
# Set significance level (alpha)
alpha = 0.05

# Interpret the result based on the p-value
if p_value > alpha:
    print('Fail to reject the null hypothesis (H0). No strong evidence of heteroscedasticity. Homoscedasticity is validated.')
else:
    print('Reject the null hypothesis (H0). Strong evidence of heteroscedasticity. Homoscedasticity is violated.')
```

> **Interpretation of the Goldfeld-Quandt Test Results**

  - **Test Statistic:** 1.0044  
    The Goldfeld-Quandt test statistic compares the variances of the residuals from two groups of data. A value close to 1 suggests similar variances in both groups, supporting homoscedasticity.

  - **p-value:** 0.4887  
    The p-value represents the probability of observing a test statistic as extreme as 1.0044 under the null hypothesis (H₀). Since **p-value = 0.4887 > 0.05**, we fail to reject the null hypothesis.

> **Conclusion**

  - **Fail to Reject H₀:** There is no strong evidence of heteroscedasticity.  
  - **Inference:** The residuals exhibit constant variance across predicted values, validating the assumption of homoscedasticity in the model.

```python
# Helper function to plot individual metric comparison with model names
def plot_metric_comparison(ax, base_metric, updated_metric, alpha_values, metric_name, ylabel, base_model_name, updated_model_name):
    """
    Plots the comparison of a single metric between the base model and the updated model (e.g., Ridge, Lasso, etc.).

    Parameters:
    ax (matplotlib.axes.Axes): The axis object on which the plot will be drawn.
    base_metric (float): The metric value for the base model (e.g., OLS).
    updated_metric (list of float): List of metric values for the updated model (e.g., Ridge).
    alpha_values (list of float): The list of alpha values used for the updated model (e.g., for Ridge or Lasso).
    metric_name (str): The name of the metric being compared (e.g., 'MAE', 'RMSE', 'R²').
    ylabel (str): The label for the y-axis to describe the metric (e.g., 'Mean Absolute Error', 'R-squared').
    base_model_name (str): The name of the base model (e.g., 'OLS Model').
    updated_model_name (str): The name of the updated model (e.g., 'Ridge Model').

    This function plots a comparison of the base and updated model's metric over the different alpha values,
    adding a vertical line to indicate the base model's metric.
    """
    # Plot the metric for the updated model across different alpha values
    ax.plot([0] + alpha_values, [base_metric] + updated_metric, marker='o', label=f'{updated_model_name} {metric_name}', color='blue')

    # Plot a vertical line for the base model's metric at alpha=0
    ax.axvline(x=0, color='red', linestyle='--', label=f'{base_model_name} {metric_name}')

    # Set the title and labels for the plot
    ax.set_title(f'{metric_name} Comparison')
    ax.set_xlabel('Alpha')  # X-axis label represents the alpha values
    ax.set_ylabel(ylabel)   # Y-axis label corresponds to the metric (e.g., MAE, RMSE, R²)

    # Add a legend to differentiate the base and updated model metrics
    ax.legend()
```

```python
# Function to compare base model and updated model (e.g., Ridge, Lasso, etc.)
def plot_comparison(base_model_metrics, updated_model_metrics, alpha_values, base_model_name, updated_model_name):
    """
    Plots the comparison of multiple metrics (MAE, RMSE, R², Adjusted R²) between the base model and the updated model.

    Parameters:
    base_model_metrics (list of float): List of metrics for the base model (e.g., OLS).
    updated_model_metrics (list of list of float): List of metrics for the updated model (e.g., Ridge or Lasso).
    alpha_values (list of float): The list of alpha values used for the updated model.
    base_model_name (str): The name of the base model (e.g., 'OLS').
    updated_model_name (str): The name of the updated model (e.g., 'Ridge').

    This function uses the helper function to plot the comparison of each of the four metrics (MAE, RMSE, R², Adjusted R²)
    between the base and updated models, in a 2x2 grid.
    """
    # Unpack the metrics for the base model and updated model
    base_mae, base_rmse, base_r2, base_adj_r2 = base_model_metrics
    updated_mae, updated_rmse, updated_r2, updated_adj_r2 = updated_model_metrics

    # Set up the plots in a 2x2 grid (2 rows, 2 columns)
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    # Plot each metric using the helper function
    plot_metric_comparison(axes[0, 0], base_mae, updated_mae, alpha_values, 'MAE', 'Mean Absolute Error', base_model_name, updated_model_name)
    plot_metric_comparison(axes[0, 1], base_rmse, updated_rmse, alpha_values, 'RMSE', 'Root Mean Squared Error', base_model_name, updated_model_name)
    plot_metric_comparison(axes[1, 0], base_r2, updated_r2, alpha_values, 'R²', 'R-squared', base_model_name, updated_model_name)
    plot_metric_comparison(axes[1, 1], base_adj_r2, updated_adj_r2, alpha_values, 'Adjusted R²', 'Adjusted R-squared', base_model_name, updated_model_name)

    # Adjust layout to prevent overlap between subplots and ensure a clean view
    plt.tight_layout()

    # Display the plot
    plt.show()
```

```python
def model_metrics(X_train, y_train, X_test, y_test, alpha_values, model_type='Ridge'):
    """
    Generalizes the metric calculation for different models (Ridge, Lasso, ElasticNet, LinearRegression).

    Parameters:
    X_train (array-like): Training features.
    y_train (array-like): Training target values.
    X_test (array-like): Test features.
    y_test (array-like): Test target values.
    alpha_values (list): List of alpha values for regularization.
    model_type (str): Type of model to use ('Ridge', 'Lasso', 'ElasticNet', 'LinearRegression').

    Returns:
    tuple: Lists of metrics (MAE, RMSE, R², Adjusted R²) for each alpha value.
    """
    # Lists to store the metrics for each alpha value
    mae_list = []
    rmse_list = []
    r2_list = []
    adj_r2_list = []

    n = X_test.shape[0]  # Number of samples in the test set
    p = X_test.shape[1]  # Number of features

    # Loop through each alpha value to train and test the model
    for alpha in alpha_values:
        # Initialize the model based on the model_type
        if model_type == 'Ridge':
            model = Ridge(alpha=alpha)
        elif model_type == 'Lasso':
            model = Lasso(alpha=alpha)
        elif model_type == 'ElasticNet':
            model = ElasticNet(alpha=alpha)
        elif model_type == 'LinearRegression':
            model = LinearRegression()
        else:
            raise ValueError(f"Unsupported model type: {model_type}")

        # Train the model
        model.fit(X_train, y_train)

        # Predict on the test set
        y_pred = model.predict(X_test)

        # Calculate the metrics
        mae = mean_absolute_error(y_test, y_pred)
        rmse = mean_squared_error(y_test, y_pred, squared=False)
        r2 = r2_score(y_test, y_pred)
        adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

        # Append the metrics to respective lists
        mae_list.append(mae)
        rmse_list.append(rmse)
        r2_list.append(r2)
        adj_r2_list.append(adj_r2)

    return mae_list, rmse_list, r2_list, adj_r2_list
```

```python
def get_best_model_metrics(mae_list, rmse_list, r2_list, adj_r2_list):
    """
    Given four lists of MAE, RMSE, R², and Adjusted R² values, returns the metrics for the model
    with the highest Adjusted R².

    Parameters:
    mae_list (list): List of MAE values.
    rmse_list (list): List of RMSE values.
    r2_list (list): List of R² values.
    adj_r2_list (list): List of Adjusted R² values.

    Returns:
    dict: Dictionary with the best model's metrics (MAE, RMSE, R², Adjusted R²).
    """
    # Find the index of the maximum Adjusted R²
    best_index = adj_r2_list.index(max(adj_r2_list))

    # Get the corresponding metrics using the best_index
    best_metrics = {
        'MAE': mae_list[best_index],
        'RMSE': rmse_list[best_index],
        'R²': r2_list[best_index],
        'Adjusted R²': adj_r2_list[best_index]
    }

    return best_metrics
```
